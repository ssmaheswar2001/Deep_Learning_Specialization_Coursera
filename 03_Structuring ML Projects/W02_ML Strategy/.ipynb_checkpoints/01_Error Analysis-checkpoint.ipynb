# Carrying Out Error Analysis
- If we're trying to get a learning algorithm to do a task humans can do and if our learning algorithm is not yet at the performance of a human. Then manually examining mistakes that our algorithm is making, can give us insights into what to do next. This process is called error analysis.
- Error analysis procedure that can let us very quickly tell whether or not his could be worth our effort.
    1. Get about, say 100 mislabeled dev set examples.
    2. Then examine them manually. Just count them up one at a time, to see how many of these mislabeled examples in our dev set are actualy pictures of dogs.
- The error analysis evaluates whether or not a idea is worth working on or not.
- 
- During error analysis, we're just looking at dev set examples that our algorithm has misrecognized.
- The conclusiong of the matrix process gives us an estimate of how worthwhile it might be to work on each of the different categories of error.
- 
- To summarize, 
    - To carry out errpr analysis, we should find a set of mislabeled examples, either in our dev set.
    - Look at the mislabeled examples for false positives and false negatives.
    - Count up the number of errors that fall into various different categories.
    - During this process, we might be inspired to generate new categories of errors.
    - By counting up the fraction of examples that are mislabeled in different ways, often this will help us prioritize or give us inspiration for new directions to go in.

# Cleaning Up incorrectly Labeled Data
- The data for our supervised learning problem comprises input X and output labels Y.
- What is we going through our data and we find that some of these output labels Y are incorrect, we have data which is incorrectly labeled?
- If the errors are reasonably random, then it's probably okay to just leave the errors as they are not spend too much time fixing them.
- 
- The incorrectly labeled examples in our dev set or test set? If we're worried about the impact of incorrectly labeled examples on our dev set or test set, the recommended way is to do during error analysis to add one extra column so that we can also count up the number of examples where the label Y was incorrect.
- 
- Few guidelines or principles to consider.
    - Apply same process to out dev and test sets to make sure they continue to come from the same distribution. The dev set is telling us where to aim to target and when we hit it, we want to generalize to the test set. If we're going to fix something on the dev set, apply the same process to the test set to make sure that they continue to come fromt the same distribution.
    - Consider examining examples our algorithm got right as well as ones it got wrong.
    - If we go into a dev and test data to correct some of the labels there, we may or may not decide to go and apply the same process for the training set.
