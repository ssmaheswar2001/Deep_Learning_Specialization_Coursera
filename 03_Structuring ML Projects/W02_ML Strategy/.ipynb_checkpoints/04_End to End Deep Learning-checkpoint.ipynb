# What is End-to-End Deep Learning ?
- 
- 
- End-to-End deep learning work reall well and it can really simplify the system and not require us to build so many hand-designed individual components.

# Whether to use End-to-End Deep Learning
- Pros of End-to-End learning :
    1. Let the data speak. If we have enough X, Y data then whatever is the most appropriate function mapping from X to Y, if we train a big enough neural network, the neural network will figure it out and by having a pure machine learning approach, our neural network learning input from X to Y may be more able to capture whatever statistics are in the data, rather than being forced to reflect human preconceptions.
    2. Less hand-designing of components needed.
- Cons of End-to-End learning
    1. May need large amount of data.
    2. Excludes potentially useful hand-designed components.
