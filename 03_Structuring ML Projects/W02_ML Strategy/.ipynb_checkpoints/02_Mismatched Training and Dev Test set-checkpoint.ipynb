# Bias and Variance with Mismatched Data Distribution
- Estimating the bias and variance of our learning algorithm helps us prioritize what to work on next. 
- The way we analyze bias and variance changes when our training set comes from a different distribution than our dev and test sets.
- If our dev data come from the same distribution as our training set, we would say that we have a large vairance problem, that our algorithm's just not generalize well from the training set which it's doing well on to the dev set, which is suddently doing much worse on.
- But in the setting where our training data and our dev data comes from a different distribution, we can no longer draw this conclusion.
- The problem with this analysis is that when we went from the training error to the dev error, 2 thinngs changed at a time
    1. The algorithm saw data in the training set but not in the dev set.
    2. The distribution of data in the dev set is different.
- In order to tease out these 2 effects it will be useful to define a new piece of data which we'll call the training-dev set.
- This training-dev set which we carve out that should have the same distribution as training sets, but we don't explicitly train our neural network on this.
- To carry our error analysis, what we should do is now look at the error of our classifier on the training set, on the training-dev set, as well as on the dev set.
- The training-dev error was measured on data that comes from the same distribution as our training set. So even though our neural network does well in a training set, it's just not generalize well to data in the training-dev set which comes from the same distribution, but it's not generalizing well to data from the same distribution that is hasn't seen before. So this example have a variance problem.
- When we went from training data that we've seen to the training-dev data that the neural network has not seen, the error increases only a little bit, but then it really jumps when we go the dev set. So this is a data mismatch problem, where data mismatched. This is a data mismatch problem, becuase our learning algorithm was not trained explicitly on data from training-dev or dev, but these 2 data sets come from different distributions. But whatever algorithm it's learning, it works great on training-dev but it doesn't work well on dev. So somehow our algorithm has learned to do well on a different distribution than what we really care about, so we call that a data mismatch problem.

# Addressing Data Mismatch
- If our training set comes from a different distribution, than our dev and test set, and if error analysis shows us that we have a data mismatch problem.
- Carry out manual error analysis and try to understand the differences b/w the training set and the dev/test sets. To avoid overfitting the test set, technically for error analysis, we should manually only look at a dev set and not at a test set.
- Make training data more similar; or collect more data similar to dev/test sets.
- If our goal is to make the training data more similar to our dev set. One techniques we can use is artificial data synthesis.
- 
- Trhough artificial data synthesis, we might be able to quickly create more data without needing to go out there and collect tons of data.
- When we're synthesizing data from a very small subset of the space, and the neural network might be overfitting.
- 
- To summarize,
    - If we think we have a data mismatch problem. We do error analysis, or look at the training ser or look at the dev set to try to figure out, to try to gain insight into how these 2 distributions of data might differ. 
    - Then see if we can find some ways to get more training data that looks a bit more like our dev set. One of the ways is artificial data synthesis.
    - But, if we're using artifical data synthesis, be cautious whether or not we might be accidentally simulate data only from a tiny subset of the space of all possible examples.
