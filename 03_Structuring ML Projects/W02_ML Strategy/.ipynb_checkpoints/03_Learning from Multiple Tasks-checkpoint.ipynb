# Transfer Learning
- We can take knowledge the neural network has learned from one task and apply that knowledge to a spearate task.
- For example, maybe we could have the neural network learn to recognize objects like cats and then use that knowledge or use part of that knowledge to help us do a better job reading x-ray scans. This is called Transfer learning
- Let's say we've trained our neural network and train it on X->Y pairs, where X is an image and Y is some object.
    - If we want to take this neural network and adapt, or we say transfer, what is learned to a different task, such as radiology diagnosis.
    - What we can do is take the last output layer of the neural network and just delete that and delete also the weights feeding into that last output layer and create a new set of randomly initialized weights just for the last layer and have that now output radiology diagnosis. 
    - So to be concrete, during the first phase of training when we're training on an image recognization task, we train all of the usual parameters for the neural network, all weights, all the layers and we have something that now learns to make recognition predictions. 
    - Having tranined that neural network, what we now do to implement transfer learning is swap in a new data set X Y, where now these are radiology images and Y are the diagnoses we want to predict and we initialize the last layer's weights.
    - So now retrain the neural network on the new data set, on the new radiology data set. 
    - We have a couple options of how we retrain the neural network with radiology data. 
        1. If we have a small radiology dataset, we might want to retrain the weights of the last layer and keep the rest of the parameters fixed.
        2. If we have enough data, we could also retrain all the layers of the rest of the neural network.
    - If we retrain all the parameters in the neural network, then this initial phase of training on image recognition is sometimes called pre-training, becuase we're using image recognitions data to pre-initialize or pre train the weight of the neural network.
    - Then if we are updating all the weights afterward, then training on the radiology data called tuning.
    - So, what we've done is, we've taken knowledge learned from mage recognition and applied it or transferred it to radiology diagnosis.
- The reason this can be helpful is that a lot of the lowe level features such as detecting edges, detecting curves, detecting positive objects, learning from that, from a very large image recognition data set, might help our learning algorithm do better in radiology diagnosis.
- It's just learned a lot about the structure and the nature of how images look like and some of that knowledge will be useful. So having learned to recognize images, it migh have learned enough about we know, just waht parts of different images look like, that knowledge about lines, dots, curves and so on, maybe small parts of objects, that knowledge could help our radiology diagnosis network learn a bit faster or learn with less data.
- We might take out the last layer of the neural network and create a new output node. Another thing we could do is actually create not just a single new output, but acutally we can create several new layers to our neural network to try predict the labels Y.
- Transfer learning makes sense when we have a lot of data for the problem we're tranasferring from and usually relatively less data for the problem we're transferring to.
- 
- When does transfer learning make sense : 
    1. If we're trying to learn from some Task A and transfer some of the knowledge to some Task B, then transfer learning makes sense when Task A and B have the same input X. 
    2. It tends to make sense when we have a lot more data for Task A than for Task B.
    3. When low level features from Task A could be helpful for learning Task B.

# Multi-task Learning
- In transfer learning, we have a sequential process where we learn from task A and then transfer that to task B. In multi-task learning, we start off simultaneously, trying to have one neural network do several things at the same time and then each of these task helps hopefully all of the other task.
- 
- We're doing is building a single neural network that is looking at each image and basicaly solving 4 probkems. It's trying to tell us does each image have each of these 4 objects in it. So one image can have multiple labels. If we train a neural network to minimize this cost function, we are carrying out multi-task learning.
- One other thing we could have done is just train 4 seprate neural networks, instead of train one network to do 4 things. But if some of the earlier features in neural network can be shared b/w these different types of objects, then we find that training one neural network to do 4 things results in better performance than training 4 completely seperate neural networks to do the 4 tasks separately.
- 
- When does multi-task learning makes sense when 3 things are true :
    1. If our training on a set of tasks that could benefit from having shared low-level features.
    2. Usually : Amount of data wehave for each task is quite similar.
    3. Can train a big enough neural network to do well on all the tasks.
- So multi-task learning enables us to train one neural network to do many tasks and this can give us better performance than if we were to do the tasks in isolation.
- Computer vision object detection, where we do see a lot of applicaton of training a neural network to detect lots of different objects.
