# Single Number Evaluation Metric
- Whenever we're tuning hyperparameters, or trying out different ideas for learning algorithms, or just trying out different options for building our machine learning system. We'll find that our progress will be much faster if we have a single real number evaluation metric that lets us quickly tell if the new thing we just tried is working better or worse than our last idea.
- Before that applied machine learning is a very empirical process. We often have an idea, code it up, run the experimenet to see how it did, and then use the outcome of the experiment to refine our ideas and keep going around the loop as we keep on imporoving our algorithm.
- One reasonable way to evaluate the performance of our classifier is to look at its precision and recall.
- Precision is, of examples that we classifier recognized as cats. If classifier A has 95% precision, this means that when classifier A says somethin is a cat, there's a 95% chance it really is cat.
- Recall is, of all the images that really are cats, what percentage were correctly recognized by our classifier? So what percentage of actual cats, are correctly recognized? So if classider A is 90% recall, this means that of all the images in, our dev sets really are cats, classifier A accurately pulled out 90% of them. 
- The problem with using precision and recall as our evaluation metric is that if classifier A does better on recall, which it does here, the classifier B does better on precision, then we're not sure which classifier is better.
- Rather than using 2 numbers, precision and recall, to pick a classifier, we just have to find a new evalutation metric that combines precision and recall.
- The standard way to combine precision and recall is something called an F1 score. 
- Evaluation metric allows us to quickly tell if classifier A or classifier B is bstter and therefore having a dev set plus single number evaluation metric distance to speed up iterating. It speeds up the iterative process of imporving our machine learning algorithm.

# Train/Dev/Test Distribution
- The workflow in machine learning is that we try a lot of different models on the training set, and then use the dev set to evaluate the different ideas and pick one and keep innovating to improve dev set performance until, finally, we have one clause that we're happy with that we then evaluate on our test set.
- 
- Choose a dev set and test set to reflect data we expect to get in future and consider important to do well on.
- The dev set and test set should come from the same distribution.
- Setting up the dev set and as well as the evaluation metric, is really defining what target we want to aim at.
- Setting the dev set and the test set to the same distrubution, we're really aiming at whatever target we hope our machine learning team will hit. The way we choose our training set will affect how well we can actually hit that target.
