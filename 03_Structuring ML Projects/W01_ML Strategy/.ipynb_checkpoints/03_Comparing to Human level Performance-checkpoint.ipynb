# Why Human-level Performance?
- As we keep training the algorithm, maybe bigger and bigger models on more and more data, the performance approaches but never surpasses some theoretical limit, which is called the Bayes optimal error.
- The Bayes optimal error is the best possible error.
- 
- 

# Avoidable Bias
- We don't actually want to do too well and knowing what human level performance is, can tell us exactly how well but not too well we want our algorithm to do on the training set.
- The difference b/w Bayes error or approximation of Bayes error and the training error to be the avoidable bias.
- 

# Understanding Human-level Performance
- 
- 
- The difference b/w out estimate of Bayes error tells us how much avoidable bias is a problem, how much avoidable bias there is.
- The difference b/w training error and dev error, that tells us how much variance is a problem, whether our algorithm's able to generalize from the training set to the dev set.
- 
- Having an estimate of human-level performance gives us an estimate of Bayes error and this allows us to more quickly make decisions as to whether we should focus on trying to reduce a bias or trying to reduce the variance of our algorithm.

# Improving our model performance
- Getting a supervised learning algorithm to work well means fundamentally hoping or assuming they can do 2 things.
    1. We can fit the trainig set pretty well, and we can think of this as that we can achieve low avoidable bias.
    2. The training set performance generalized pretty well to the dev/test set and this is sort of saying that variance is not too bad.
- If we want to improve the performance of our ML system, looking at the difference b/w our training error and our proxy for Bayes error and just gives us a sense of the avoidable bias. In other words, just how much better do we think we should be trying to do on our training set. 
- Then look at the difference b/w our dev error and our training error as an estimate of how much of a variance problem we have. In other words, how much harder we should be working to make our performance generalized from the training set to the dev set that if it wasn't trained on explicitly.
