# Softmax Regression
- Binary classification, where we had 2 possible labels 0 or 1.
- If we have multiple possible classses, there's a generalization of logistic regression called softmax regression.
- The notation we're going to use is C to denote the number of classes we're categorize our input into.
- The number of units upper layer which is layer L is going to equal to C.
- What we want is for the number of units in the upper layer to tell us what is the probability of each of these classes. 
- 
- The standard model for getting our network to do this uses what's called a softmax layer, and the output layer in order to generate the outputs. 
- Previusly, our activation functions used to take in a single row value input. For example, the sigmoid and the value activation funtions input the real number and output a real number,
- The unusual thing about the softmax activation function is, because it needs to normalized across the different possible outputs and needs to take a vector and puts in outputs of vector.
- 
- We can learn more complex non-linear decision boundaries to seperate out multiple different classes.
