# Why does Batch Norm work?
- We've seen how normalizing the input features, the X's, to mean zero and variance one, how that speed up learning. So rather than having some features that range from zero to one, and some from one to a 1000, bu normalizing all the features, input features X, to take a similar range of values that can speed up learning. So, one intuition behind why batch norm works is, this is doing a similar thing, but further values in our hidden units and not for out input there.
- The second reason why batch norm works, is it makes weights, later or deeper than our network, say the weight on layer 10, more robust to changes to weights in earlier layers of the neural network. say in layer one.
- The idea of our data distribution changing, called covariate shift and the idea is that id we've learned some X to Y mapping, if the distribution of X changes, then we might need to retrain our learning algorithm and this is true even if the function, the ground true function, mapping from X to Y, remains unchanged.
- 
- What batch norm does, is it reduces the amount that the distribution of these hidden unit values shifts around.and indeed they will change when the neural network updates the parameters in the earlier layers. But what batch norm ensures is that no matter how it changes, the mean and vairance of Zs will remain the same. 
- Batch norm limits the amount to which updating the parameters in the earlier layer can affect the distribution of values. So batch norm reduces the problem of the input values changing, it really causes these values to become more stable, sot that the later layers of the neural network has more firm ground to stand on and even though the input distribution changes a bit, it changes less and what this does is, even as the earlier layers keep learning, the amount that this forces tha layers to adapt to early as layer changes is reduced or, if we will, it weakens the coupling b/w what the early layers parameters has to do and what the later layers parameters have to do and so it allows each layer of the network to learn by itself, a little bit more independently of other layers, and this has the effect of speeding up of learning in the whole network.
- So batch norm especially from the perspective of one of the later layers of the neural network, the earlier layers don't get to shift around as much, because they're constrained to have the same mean and variance. So this makes the job of learning on the later layers easier.
