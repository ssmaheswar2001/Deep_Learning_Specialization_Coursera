# Normalizating Inputs
- When trainintg a neural network, one of the techniques to speed up our training is if we normalize our inputs.
- Normalizing our inputs corresponds to 2 steps : 
    1. To subtract out or to zero out the mean. This means that we just move the training set until it has zero mean.
    2. To normalize the variance
    - If we use this to scale our training data, then use the same mu and sigma to normalize our test set because we want out data both training and test examples to go throuh the same transformation defined by the same Mu and Sigma**2 calculated on our training data.
- 
- Why do we want to normalize the input features?
    - That the cost function turnsout that if we use unnormalized input features, it's more likely that our cost function will look like a very squished out bar, very elongated cost function where the minimum we're trying to find.
    - But if our features are on very different scales, say feature x1 ranges from 1-1000 and the feature x2 ranges from from 0-1 then it turns out that the ratio or the range of values for the parameter w1 and w2 will end up taking on very different values. If we are running gradient descent on a cost function, then we might have to use a very small learning rate because the gradient descent migh need a lot of steps to oscillate back and forth before it finally finds its way to the minimum. 
    - But if normalize the features, then our cost function wil on average look more symmetric. If we have more spherical, then wherever we start, gradient descent can pretty much go straight to the minimum. We can take much larger steps where gradient descent need, rather than needing to oscillate around. The cost function will be in a more round and easier to optimize when we're features are on similar scales.

# Vanishing / Exploding Gradients
- One of the problems of training neural networks, especially very deep neural networks, is data vanishing and exploding gradients. This means that when we're training a very deep network our derivatrives or our slopes can sometimes get either very big or every small, maybe even exponentially small, and this makes training difficult.
- Careful choices of the random weight initialization significantly reduces the problem.
- 
- If our activations or gradients increase or decrease exponentially as a function of L, then these values could get really big or really small and this makes training difficult.
