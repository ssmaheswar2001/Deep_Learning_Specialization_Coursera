# Regularization
## Logistic Regression
- If we suspect our neural network is overfitting our data, we have a high variance problem, one of the 1st things we should try is probably regularization.
- Adding regularization will often help to prevent overfitting, or to reduce variance in our network.
- To add regularization to logistic regression, add lambda, which is called the regularization parameter
- A square Euclidean norm of the parmater vector W, and this is called L2 regularization.
- Why do we regularize just the parameter w? because if we look at our parameters, W is usually a high dimensional parameter vector, especialy with a high variance problem. W just has a lot of parameters, so we aren't fitting all the parameters well, whereas b is just a single number.
- If we use L1 regularization, then W will end up being sparse. What that means is that the w vector will have a lot of zeros in it. This can help with compressing the model, because the set of parameters are zero, then we need less memory to store the model.
- L2 regularization is more often used.
- Lambda is another hyperparameter

- Let's say we want network with 3 layers l = 3. 
- We'll illustrate how to represent dropout in a single layer. 
- Set a vector d and d32 is going to be the dropout vector for the layer 3 and this is going to be the same shape as a3.
- When we see a number which is less than keep.prob and there will be the probability that a given hidden unit will be kept. If keep.prpb = 0.8, then this means that there's a 0.2 chance of eliminating any hidden unit.
- Therefore, each example for each hidden unit there's a 0.8 chance that corresponding d3 will be one. and a 20% chance that the corresponding d3 will be one and a 20% hance there will be zero.
- So this random number being less than 0.8 it has a 0.8 chance of being one or be true, and 20% or 0.2 chance of being false of being zero.
- Then finally, we're going to take a3 and scale it up by dividing by 0.8 or really dividng by our keep.prob parameter. In order to not reduce the expected value of z, what we do is we need to divide it by 0.8 becuase this will correct or just a bump that back up by roughly 20% that we need. So it's not changed the expected value of a3. This line here is what's called the inverted dropout technique. 
- This inverted dropout technique by divindg by the keep.prob, it ensures that the expected value of a3 remains the same.

# Understanding Dropout
## Why does drop-out work?
- Intuition : Can't rely on any one feature, so have to spread out weights.
- Drop out randomly knocks out units in our network. so it's as if one every iteration we're working with a smaller neural network. So using a smaller neural network seems like it should have a regularizing effect.
- The effect of implementing dropout is that its strength the ways and similar to L2 regularization, it helps to prevent overfitting, but it turns out that dropout can formally be shown to be an adapative form of L2 regularization, bu the L2 penalty on different ways are different depending on the size of the activation is being multiplied into that way. 
- One of the practice we have to choose was the keep prop which is a chance of keeping a unit in each layer. So it is also feasible to vary keep-propped by layer.
- For layers, where we're more worried about overfitting really the layers with a lot of parameters we could say keep prop to be smaller to apply a more powerful form of dropout.
- If we're worried about some layers of fitting than others, we can set a lower key prop for some layers than others.
- One other alternative might be to have some layers where we apply dropout and some layers where we don't apply drop out and then just have one hyper parameter which is a key prop for the layers for which we do apply drop out.
- First successful implemenations of dropouts were to computer vision, so in computer vision, the input sizes so big in putting all these pixels that we almost never have enough data.
- So dropout is a regularization technique which helps prevent overfitting.
- One downside of dropout is that the cost function J is no longer well defined on every iteration.
