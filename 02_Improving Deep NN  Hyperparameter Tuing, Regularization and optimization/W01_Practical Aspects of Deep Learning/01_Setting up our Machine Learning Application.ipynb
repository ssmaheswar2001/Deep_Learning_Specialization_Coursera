# Train / Dev / Test sets
- Making good choices in how we set up our training, development, and test sets can make a hige difference in helping us quickly find a good high-performance neural network.
- Wehn training a neural network, we have to make a lot of decisions, such as
    - How many layers will our neural network have?
    - How many hidden units do we want each layer to have?
    - What's the learning rate?
    - What are the activation functions we want to use for the different layers?
- Applied machine learning is a highly iterative process, in which 
    - we often start with an idea, such as build a neural network of a certain number of layers, a certain number of hidden units and so on. 
    - Then we just have to code it up and try it, by running our code. 
    - We run an experiment and we get back a result that tells us how well this particular network or this particular configuration works. Based on the outcome, we might then refine our ideas and change our choices and maybe keep iterating, inorder to try to find a better and a better, neural network.
- One of the things that determine how quickly we can make progress is how efficiently we can go aroung the cycle. 
- 
- Setting up our data sets well, in terms of our train, development and tests sets can make us much more efficient.
- It was common practice to take all our data and split it according to 70/30 in terms of train/ test. In terms dev, 60/20/20% split.
- Having set up a train dev and test set will allow us to intergate more quickly. It will also allow us to more efficiently measure the bia and variance of our algorithm, so we can more efficiently select ways to improve our algorithm.
