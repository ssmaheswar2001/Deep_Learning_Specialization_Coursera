# Logistic Regression
- This is a learning algorithm that we use when the output label Y in a supervised learning problem are all either zero or one.
- Given an input feature vector X maybe corresponding to an image that we want to recognize as either a cat or a not a cat, we want an algorithm that can output a prediction (y_hat), which is our estimate of Y.
- We want Y_hat to be the probability of the chance that Y is equal to 1 given the input feature X.
- Given an input X and the parameter W and b, we generate the output Y_hat.
- When we implement logistic regression, our job is to try to learn parameters W and B so that Y_hat becomes a good estimate of the chance of Y = 1.
-
