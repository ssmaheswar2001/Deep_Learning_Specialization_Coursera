# Transfer Learning
- If we're building a computer vision application rather than training the ways from scratch, from random initialization, we often make much faster progress if we download ways that someone else has already trained on the network architecture and use that as pre-training and transfer that to a new task that we might be interested in.
- One of the trick that could speed up training is we just pre-compute the layer, the features of re-activations from the layer and just save them to disk
- The advatnage of the safety disk or the pre-compute method is that we don't need to recompute those activations everytime we take a epoch or take through a training set.
- One rule of thumb is if we have a larger label data, one thing we could do is then freeze fewer layers and then train these later layers.
- If we have more data, the number of layers we've freeze could be smaller and then the number of layers we train on top could be greater.
- The idea is that if we pick a dataset and maybe have enough data not just to train a single softmax unit but to train some other size neural network that comprises the last few layers of the final network that we end up using. 
- If we have a lot of data, one thing we might do is take the open source network and ways and use the whole thing just as initialization and train the whole network.

# Data Augmentation
- Most computer vision task could use more data. So data augmentation is one of the techniques that is often used to improve the performance of computer vision systems.
- The simplest data augmentation method is mirroring on the vertical axis.
- Another commonly used technique is random cropping. 
- 
## Color Shifting
- The 2nd type of data augmentation that is commonly used is color shifiting.
- We add R, G and B channels different distortion. 
- There are different ways to sample R, G and B. One of the ways to implement color distoration uses an algorithm called Principal Component Analysis (PCA).
- The idea of the PCA color augmentation is for example, if our image is mainly purple, if it mainly has red and blue tints, and very little green, then PCA color augmentation, will add and subtract a lot to red and blue, where it balance all the greens, so kind of keeps the overall color of the tint the same. 
- 
## Implementing distorations during training
- If we have a small training set, we can do almost anythng and we'll be okay.
- But the very training set which we have a CPU thread that is constantly loading images of our hard disk. What we can do is use maybe a CPU thread to implement the distorations, yet the random cropping, or the color shifting, or the mirroring, but for each image, we might then edn up wit some distorted version of it.
- So our CPU thread is constanly loading data as well as implementing whether the distortions are needed to form a batch or really many batches of data. This data is then constaly passed to some other thread or some other process for implementing training.
- A pretty common way of implementing data augmentation is to really have one thread, almost 4 threads, that is responsible for loading the data and implementing distortions and then passing that to some other thread or some other process that then does the training.
