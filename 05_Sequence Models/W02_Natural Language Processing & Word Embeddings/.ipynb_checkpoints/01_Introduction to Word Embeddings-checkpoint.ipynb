# Word Representation
- One of the key ideas we learn about is word embeddings, which is way of representing words.
- Through these ideas of word embeddings, we'll be able to build NLP applciations, even with models with size of relatively small label training sets.
- We've been representing words using a vocabulary of words or using one-hot vector.
- One of the weakness of the one-hot representation is that it treats each word as a thing until itself, and it doesn't allow an algorithm to easily generalize the cross words.
- 

## Featurized representation : Word Embedding
- Instead of an one-hot reprsentation we can instead learn a f4eaturized representation with each of the words. 
- For every word in the dictionary, we could learn a set of features and values for each of them.
- A lot of the features of apple and organge are actually the same or very similar values. So this increase the odds of the learning algorithm that has figured out that organge juice is a thing, to also quickly figure out that apple juice is a thing. This allows it to generalize better across different words.
- 
- So we need to learn high dimensional feature vectors like these, that gives a better representation than one-hot vectors for representing different words.

## Visualizing word embeddings
- If we're able to learn a 300 dimensional feature vector or 300 dimensional embedding for each words, one of the popular things to do it also to take this 300 dimensional data and embed it in a 2D space so that we can visualize them.
- One common algorithm is the t-SNE algorithm.
- 
- Word embeddings algorithms like this can learn similar features for concepts that feel like they should be more related and end up getting mapped to a more similar feature vectors.

## Transfer Learning & Word Embeddings
- We carry out transfer learning using word embeddings : 
    1. Learn word embeddings from a large text corpus, a very large text corpus or we can also download pre-trained word embeddings online. 
    2. We can then take these word embeddings and transfer the embedding to new task, where we have a much smaller labeled training sets.
    3. Finally, as we train our model on our new task.
- So word embeddings tend to make the biggest difference when the task we're trying to carry out has a relatively smaller training set.
- In transfger learning settings, if we're trying to transfer from task A to task B, the process of transfer learning is just most useful when we happen to have a ton of data for A and a relatively smaller data set for B.
