## Skip-grams
- Lets say we're given the sentence 'I want a glass of orange juice to go along with my cereal' in our training set. 
- In the skip-gram model, what we're  going to do is come up with a few context to target errors to create our supervised learning problem.
- We randomly pick a word to be the context word. Let's say we choose the word orange. We're going to do random pick another word withing some widow. Let's say we might pick juice to be a target word, or choose 2 words before where the target could be glass.
- We'll set up a supervised learning problem where given the context word, we're asked to predict what is a randomly chosen word within word window of that input context word.
- 
## Model
- Let's say that we are using vocab of 10,000 words. 
- The basic supervised learning problem is that we want to learn the mapping from Context c, such as the word orange to some target t, which might be the word juice or glass.
- So to represent the input such as the word orange, we can start our with some one hot vector O_c for the context words.
- Then we can take the embedding matrix E multiplied with vector O, which gives us our embedding vector for the input context word E_c.
- Then we're going to take the vector E_c and feed it to a softmax unit.
- The softmax model, calculates the probability of different target words given the input context word.
- This is called the skip-gram model because it takes input one word like orange and then trying to predict some words skipping a few words from the left or the right side. To predict what comes little bit after the context words.
