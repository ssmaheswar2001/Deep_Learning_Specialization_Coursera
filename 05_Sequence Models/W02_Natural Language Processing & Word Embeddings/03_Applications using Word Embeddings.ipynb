# Sentiment Classification
- Sentiment classification is the task of looking at a piece of text and telling someone likes or dislikes the thing they're talking about.
- One of the challenges of sentiment classification is we might not have a huge label training set for it.
- But with word embeddings, we're able to build good sentiment classifiers even with only modest-size label training sets.

## Sentiment classification problem
- The input X is a piece of text and the output Y that we want to predict is what is the sentiment, such as the star rating.
- 

## Model
- We can take a sentence like 'dessert is excellent' and look up those words in our dictionary.
- Given these 4 words and look up the one-hot vector multiplied by the embedding matrix E, which can learn from a much larger text corpus and use that to extract out the embedding vector for the words.
- One wat to build a classifer is that we can take these vectors, let's say these are 300-D vectors, and we could then just sum or average them. Then pass to a soft-max classifier which then outputs Y_hat. The softmax can output what are the probabilities of the 5 possible outcomes from 1-5 star.
- One of the problems with the algorithm is it ignores word order. This is a very nagative review 'Completely lacking in good taste, good service and good ambiance'. But the word good appears a lot. So ifwe use an algorithm it ignores word order and just sums or averages all of the embeddings for the different words, then we end up having a lot of the representation of good in our final feature vector and our classifier will probably think this is a good review even though this is actually very harsh.
- 

## RNN for sentiment classification
- So instead of just summing all our word embeddings, we can instead use a RNN for sentiment classification.
- The job of the RNN is to then compute the representation at the last time step that allows us to predict Y_hat.

## Addressing bias in word embeddings
1. Identify the direction corresponding to a particular bias we want to reduce or eliminate.
    - For the case of gender, what we can do is take the embedding vector for he - embedding vector for she. and average them.
    - This will allows us to figure out the gender direction or the bias direction.
2. Neutralize : For every word that is not definitional, project to get rid of bias. Word like doctor and babysitter that want to be gender neutral.
    - Project them onto to the axis to reduce their components or to eliminate their component, in the bias direction 
3. Equalize pairs : Pairs of words such as grandmother and grandfather, or girl and boy, where we want the only difference in their embedding to be the gender. We'd like to do is to make sure that words like grandmother and grandfather are both exactly the same similarity or exactly the same distance, from words that should be gender neutral, such as babysitter or such as doctor.
