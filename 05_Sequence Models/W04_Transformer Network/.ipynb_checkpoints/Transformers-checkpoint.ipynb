# Transformer Network Intuition
## Transformer Network Motivation
- As the complexity of our sequence task increases, so does the complexity of our model.
- As the complexity of our sequence task increase, so does the complexity of our model. 
    - We started with RNN and found that it had some problems with vanishing gradients, which made it hard to capture long range dependencies and sequences. We then looked at the GRU and then the LSTM model as a way to resolve many of those problems where we make use of gates to control the flow of information. 
- All of these models are still sequential models in that they ingested the input, maybe the input sentence one word or one token at the time. So as if each unit was like a bottleneck to flow of information.
- 
- The transformer architecture, which allows us to run a lot more of these computations for an entire sequence in parallel. So we can ingest an entire sentence all at the same time, rather than just processing it one word at a time from left to right.
- The major innovation of the transformer architecture is combining the use of attention based representations and a CNN convolutional neural network style of processing.
- To understand the attention network, there will be 2 key ideas will go through :
    1. Self attention
        - The goal of self attention is, if we have a sentence of 5 words will end up computing 5 representations for these 5 words A1, A2, A3, A4, A5, this will be an attention based way of computing representation for all the words in our sentence in parallel.
    2. Multi-Head Attention
        - The multi head attention is basically for loop over the self attention process. So we end up with multiple versions of these representations. It turns out that these representations, which will be very rich representations, can be used for machine translation or NLP tasks to great effectiveness.
- 

# Self Attention
- We've seen how attention is used with sequential neural networks such as RNNs. To use attention with a style more like CNNs, we need to calculate self-attention, where we create attention-based representations for each of the words in our input sentence. 
- In terms of the actual calculation, it won't be too differnt from the attention mechanism we saw previously as applied in the context of RNNs except we'll compute these representations in parallel for all 5 words in a setence. 
- The main different is that for every word, we have 3 values called the query, key and value. These vectors are the key inputs to the computing the attention value for each words. 
- 
- These matrices W^Q, W^K and W^V are parameters of the learning algorithm, and they allow us to pull of these query, key and value vectors for each word.
- The intuition behind the intent of the query, key and value of vectors are : 
    - Q3 is a question that we get to ask about 'l'Afrique', maybe like what's happening there? 
    - We compute the inner product b/w Q3 and K1 and this will tell us how good is an answer where it's one to the questionb of that's happening in Africa.
    - Then we compute the inner product b/w Q3 and K2 and this is intended to tell us how good is viste an answer to the question of what's happening in Africa and so on for the other words in the sequence.
    - The goal is this operation is to pull up the most information that's needed to help us compute the most useful representation A3
    - Then finally we're going to take these softmax values and multiply them with v's.
    - Finally, we sum up them all up.
- We could use this process for all 5 words in our sequence to get similarly rich representation. If we put all of these 5 computations together     
- 
- To recap, associated with each of the 5 words we end up with a query, a key and a value.
    - The query lets us ask a question about the word, such as what's happening in Africa. 
    - The key looks at all the other words, and by the similarity to the query helps us figure out which words gives the most relevant answer to that question. 

# Multi-Head Attention
- A big for loop over the self attention mechanism.
- Each time we calculate self attention for a sequence is called a head.
- We got the vectors Q, K and V for each of the input terms by multiplying them by a few matrices. With multi head attention, we take same set of query key and value vectors as inputs. 
- We're going to do the self-attention handlful of times.
- We compute these different heads values in parallel because no one has value depends on the value of any other head.
